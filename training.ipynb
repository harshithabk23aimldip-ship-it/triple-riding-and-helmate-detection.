```python
# === STEP 2: Import packages ===
from roboflow import Roboflow
from ultralytics import YOLO
import torch

# === STEP 3: Connect to Roboflow and download your dataset ===
# Replace YOUR_API_KEY with your Roboflow API key
rf = Roboflow(api_key="IwGHi2o2pMoDUwGWinns")
project = rf.workspace("harshas-workspace").project("triple-ride-detection")
dataset = project.version(1).download("yolov8")
# === STEP 4: Train YOLOv8 model ===
model = YOLO("yolov8n.pt")     # You can change to yolov8s.pt for higher accuracy
model.train(
    data=dataset.location + "/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8
)

# === STEP 5: Evaluate and save weights ===
metrics = model.val()
print(metrics)

# === STEP 6: Locate best model ===
print("Training complete. Best model saved at:")
!ls runs/detect/train/weights/

# === STEP 7: OPTIONAL - Export model ===
!cp runs/detect/train/weights/best.pt /content/best.pt
```
